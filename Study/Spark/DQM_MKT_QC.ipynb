```
def traverse_dir_recur(dir,name):
    l = os.listdir(dir)
    y=max(l)
    ml=os.listdir(dir+y+"/")
    m=max(ml)
    dl=os.listdir(dir+y+"/"+m+"/")
    d=max(dl)
    path=[dir+y+"/"+m+"/"+d+"/"+name]
    return path
```

```
def periodCheck(df,n):
  marscal = csvRead("GCAL.csv","dbfs:/mnt/RAW/FILES/KIND_360_GLOBAL/CONFIG_FILES/",",")
  
  ###################################################################################################################################################################################
  marsCAL= marscal.select("GMARSWINY","GMARSWINY___T","GMARSYEAR","GMARSPINY___T","GMARSPINY","GPRDSEQ","GCALMINY___T",'GCALMINY','GMARSWINP').distinct()
  
  columns = ["GMARSWINY","GMARSYEAR","GMARSPINY", "GPRDSEQ","GCALMINY","GMARSWINP"]
  for column in columns:
    marsCAL = marsCAL.withColumn(column, col(column).cast(IntegerType()))
    
  ###################################################################################################################################################################################  
  marsCAL_day=marscal.select("GMARSWINY","GMARSWINY___T","GMARSYEAR","GMARSPINY___T","GMARSPINY","GMARSDINP","GCALDINM","GPRDSEQ","GCALMINY___T",'GCALMINY',"GCALYEAR","GCALWINY",'GMARSWINP').distinct()
  
  columns1 = ["GMARSWINY","GMARSYEAR","GMARSPINY","GCALDINM","GMARSDINP", "GPRDSEQ","GCALMINY",'GMARSWINP']
  for column in columns1:
    marsCAL_day = marsCAL_day.withColumn(column, col(column).cast(IntegerType()))
  
  
  try:
    if Market in (['AT','BE']):
      df=periodFilter(MKT_DF,"PER_SDESC")
      month_split_item=F.split(df['PER_SDESC'],' ')
      df=df.withColumn('YEAR_M',month_split_item.getItem(1))
      df=df.withColumn('Week_No',month_split_item.getItem(2))
      
      marsCAL = marsCAL.withColumn("GMARSWINY",col("GMARSWINY").cast(IntegerType())).drop("GCALMINY","GCALMINY___T").distinct()
      df=cast(df,["Week_No"],IntegerType())
      
      
      df_per = marsCAL.join(df,(df["YEAR_M"]==marsCAL["GMARSYEAR"]) & (df.Week_No == marsCAL["GMARSWINY"]))\
                      .withColumnRenamed("GMARSPINY___T","Period")\
                      .withColumnRenamed("GMARSWINP","Weekno")\
                     .withColumnRenamed("GMARSYEAR","Year")
      
      df_test=df_per.select("Year","Period","Weekno").distinct()
      df_test=df_test.withColumn("total week",F.sum("Weekno").over(Window.partitionBy("Year","Period")))\
                     .filter(col("total week")!=10)\
                     .groupBy("Year","Period")\
                     .agg(F.collect_set("Weekno").alias("Weekno"))\
                     .distinct()
                 
      error=df_test.rdd.collect()
      
    elif Market in (["SE","PT"]):
      df=periodFilter(MKT_DF,"PER_SDESC")
      df=df.withColumn('Day',substring(col("PER_SDESC"),3,2))
      df=df.withColumn('Month',substring(col("PER_SDESC"),6,2))
      df=df.withColumn('YEAR_M',concat(lit("20"),substring(col("PER_SDESC"),9,2)))
      df=cast(df,["Month","YEAR_M","Day"],IntegerType())
      
      df_per = marsCAL_day.join(df,(df.YEAR_M==marsCAL_day.GMARSYEAR) & (df.Day == marsCAL_day.GCALDINM) & (df.Month==marsCAL_day.GCALMINY))\
                          .withColumnRenamed("GMARSPINY___T","Period")\
                          .withColumnRenamed("GMARSWINP","Weekno")\
                          .withColumnRenamed("GMARSYEAR","Year")
      
      df_test=df_per.select("Year","Period","Weekno").distinct()
      df_test=df_test.withColumn("total week",F.sum("Weekno").over(Window.partitionBy("Year","Period")))\
                     .filter(col("total week")!=10)\
                     .groupBy("Year","Period")\
                     .agg(F.collect_set("Weekno").alias("Weekno"))\
                     .distinct()
                     
      error=df_test.rdd.collect()
      
    elif Market=="MX":
      df=periodFilter(MKT_DF,"PER_SDESC")
      Itemsplit_col = F.split(df['PER_SDESC'], ' ')
      df = df.withColumn('PER_SDESC', Itemsplit_col.getItem(1))
      variable1 = F.split(df['PER_SDESC'], '-')
      df = df.withColumn('Week', variable1.getItem(0))
      df = df.withColumn('YEAR_M', concat(lit('20'),variable1.getItem(1))) 
      df=cast(df,["Week","YEAR_M"],IntegerType())
      
      df_per =marsCAL.join(df, (df.Week == marsCAL.GMARSWINY) &(df.YEAR_M == marsCAL.GMARSYEAR))\
                          .withColumnRenamed("GMARSPINY___T","Period")\
                          .withColumnRenamed("GMARSWINP","Weekno")\
                          .withColumnRenamed("GMARSYEAR","Year")
      
      df_test=df_per.select("Year","Period","Weekno").distinct()
      df_test=df_test.withColumn("total week",F.sum("Weekno").over(Window.partitionBy("Year","Period")))\
                     .filter(col("total week")!=10)\
                     .groupBy("Year","Period")\
                     .agg(F.collect_set("Weekno").alias("Weekno"))\
                     .distinct()
                     
      error=df_test.rdd.collect()
      
    elif Market=="IE":
      df=periodFilter(MKT_DF,"PER_SDESC")
      date_split_item = F.split(df['PER_SDESC'], '/')
      df = df.withColumn('YEAR_M',concat(lit('20'), date_split_item.getItem(1)))
      df = df.withColumn('Temp_Week', date_split_item.getItem(0))
      Week_split_item = F.split(df['Temp_Week'], ' ')
      df = df.withColumn('Day', Week_split_item.getItem(1))
      df = df.withColumn('Month', Week_split_item.getItem(0))
      df=cast(df,["Day","YEAR_M"],IntegerType())
      
      
      df_per = marsCAL_day.join(df,(df.YEAR_M==marsCAL_day.GCALYEAR) & (df.Day == marsCAL_day.GCALDINM ) & (df.Month==marsCAL_day.GCALMINY___T) )\
                      .withColumnRenamed("GMARSPINY___T","Period")\
                      .withColumnRenamed("GMARSWINP","Weekno")\
                      .withColumnRenamed("GMARSYEAR","Year")
      
      df_test=df_per.select("Year","Period","Weekno").distinct()
      df_test=df_test.withColumn("total week",F.sum("Weekno").over(Window.partitionBy("Year","Period")))\
                     .filter(col("total week")!=10)\
                     .groupBy("Year","Period")\
                     .agg(F.collect_set("Weekno").alias("Weekno"))\
                     .distinct()
                     
      error=df_test.rdd.collect()
      
    elif Market=="ES":
      df=periodFilter(MKT_DF,"PER_SDESC")
      weekYear = F.split(df['PER_SDESC'], ' ')
      date_split_item=F.split(weekYear.getItem(1),'-')
      df = df.withColumn('YEAR_M',concat(lit('20'), date_split_item.getItem(1)))
      df = df.withColumn('Week', date_split_item.getItem(0))
      df=cast(df,["Week","YEAR_M"],IntegerType())
      
      df_per =marsCAL.join(df, (df.Week == marsCAL.GMARSWINY) &(df.YEAR_M == marsCAL.GMARSYEAR))\
                          .withColumnRenamed("GMARSPINY___T","Period")\
                          .withColumnRenamed("GMARSWINP","Weekno")\
                          .withColumnRenamed("GMARSYEAR","Year")
      
      df_test=df_per.select("Year","Period","Weekno").distinct()
      df_test=df_test.withColumn("total week",F.sum("Weekno").over(Window.partitionBy("Year","Period")))\
                     .filter(col("total week")!=10)\
                     .groupBy("Year","Period")\
                     .agg(F.collect_set("Weekno").alias("Weekno"))\
                     .distinct()
                     
      error=df_test.rdd.collect()
    
    if len(error)==0:
        report_df.set_value(n,"Market",Market)
        report_df.set_value(n,"Quality Check","Checking Period Completeness")
        report_df.set_value(n,"Status","Pass")
        report_df.set_value(n,"Error/Comment","All the periods are complete with 4 weeks present")
    else:
        report_df.set_value(n,"Market",Market)
        report_df.set_value(n,"Quality Check","Checking Period Completeness")
        report_df.set_value(n,"Status","Failed")
        report_df.set_value(n,"Error/Comment","Some periods have incomplete weeks.Periods which are incomplete:"+str(error))
  except:
    report_df.set_value(n,"Market",Market)
    report_df.set_value(n,"Quality Check","Checking Period Completeness")
    report_df.set_value(n,"Status","Pass")
    report_df.set_value(n,"Error/Comment","The market does not contain week level data")
    
```

```
def getSchema(df):
  col = []
  for j in range(len(df.columns)):
    if df.columns[j]=="MKT_SDESC":
      mkt=df.select("MKT_SDESC").distinct().rdd.flatMap(lambda x : x).collect()
      col.append((df.columns[j],df.dtypes[j][1],mkt))
    elif df.columns[j]=="PROD_LEVEL":
      lvl=df.select("PROD_LEVEL").distinct().rdd.flatMap(lambda x : x).collect()
      col.append((df.columns[j],df.dtypes[j][1],lvl))
    else:
      col.append((df.columns[j],df.dtypes[j][1],['NA']))
      
  colrdd = sc.parallelize(col)
  coldf = spark.createDataFrame(colrdd,['Column','Data type','Contents'])
  return coldf
```

```
report_df = pd.DataFrame(columns=['Market','Quality Check','Status','Error/Comment'])
```

```
def colNameQC(df,n):
    
#   DF=df.drop("MARKET")
  DF=df
  col_list=[item.strip(' ') for item in DF.columns]
  ref_col=Ref.filter(col("MARKET")==Market).select("Column").rdd.flatMap(lambda x: x).collect()
  extra=[i for i in col_list + ref_col if i not in ref_col]
  miss=[i for i in col_list + ref_col if i not in col_list]
  if len(miss)==0 & len(extra)==0:
    report_df.set_value(n,"Market",Market)
    report_df.set_value(n,"Quality Check","Columns Name Check")
    report_df.set_value(n,"Status","Pass")
    report_df.set_value(n,"Error/Comment","All columns are present with correct name")
  else:
    report_df.set_value(n,"Market",Market)
    report_df.set_value(n,"Quality Check","Columns Name Check")
    report_df.set_value(n,"Status","Failed")
    report_df.set_value(n,"Error/Comment","Column name does not match.Columns missing:"+str(miss)+" & extra are:"+str(extra))
```

```
def dataType(df,n):
  col_list=[]
  DF=df
  RefDict=Ref.filter(col("MARKET")==Market).select("Column","Data Type").toPandas().set_index("Column").to_dict()
  for i in  range(len(DF.columns)):
    try:
      if RefDict['Data Type'][DF.columns[i].strip(' ')]==DF.dtypes[i][1]:
        flag=0
      else:
        col_list.append((DF.columns[i]))
        flag=1
    except:
      col_list.append((DF.columns[i]))
      flag=1
      
  if flag==1:
    report_df.set_value(n,"Market",Market)
    report_df.set_value(n,"Quality Check","Data Type Check")
    report_df.set_value(n,"Status","Failed")
    report_df.set_value(n,"Error/Comment","Data Type Failed for columns:"+str(col_list))
  else:
    report_df.set_value(n,"Market",Market)
    report_df.set_value(n,"Quality Check","Data Type Check")
    report_df.set_value(n,"Status","Pass")
    report_df.set_value(n,"Error/Comment","Data Type Proper for all the columns present in the data")
```

```
def mkt_Long(df,n):
  try:
    mkt_list=df.select("MKT_SDESC").distinct().rdd.flatMap(lambda x: x).collect()
    ref_list=Ref.filter((col("Column")=="MKT_SDESC")&(col("MARKET")==Market)).select("Contents").distinct().rdd.flatMap(lambda x : x).collect()
    extra=[i for i in mkt_list + ref_list if i not in ref_list]
    miss=[i for i in mkt_list + ref_list if i not in mkt_list]
    if len(miss)==0 & len(extra)==0:
      report_df.set_value(n,"Market",Market)
      report_df.set_value(n,"Quality Check","MKT_SDESC values check")
      report_df.set_value(n,"Status","Pass")
      report_df.set_value(n,"Error/Comment","All markets channels present")
    else:
      report_df.set_value(n,"Market",Market)
      report_df.set_value(n,"Quality Check","MKT_SDESC values check")
      report_df.set_value(n,"Status","Failed")
      report_df.set_value(n,"Error/Comment","Market channels does not match.Market channels missing:"+str(miss)+" & extra are:"+str(extra))
  except:
    report_df.set_value(n,"Market",Market)
    report_df.set_value(n,"Quality Check","MKT_SDESC values check")
    report_df.set_value(n,"Status","Pass")
    report_df.set_value(n,"Error/Comment","No column named MKT_SDESC")
    
```

```
def prod_Lvl(df,n):
  
  try:
    mkt_list=df.select("PROD_LEVEL").distinct().rdd.flatMap(lambda x: x).collect()
    ref_list=Ref.filter((col("Column")=="PROD_LEVEL")&(col("MARKET")==Market)).select("Contents").distinct().rdd.flatMap(lambda x : x).collect()
    extra=[i for i in mkt_list + ref_list if i not in ref_list]
    miss=[i for i in mkt_list + ref_list if i not in mkt_list]
    if len(miss)==0 & len(extra)==0:
      report_df.set_value(n,"Market",Market)
      report_df.set_value(n,"Quality Check","PROD_LEVEL values check")
      report_df.set_value(n,"Status","Pass")
      report_df.set_value(n,"Error/Comment","All levels present")
    else:
      report_df.set_value(n,"Market",Market)
      report_df.set_value(n,"Quality Check","PROD_LEVEL values check")
      report_df.set_value(n,"Status","Failed")
      report_df.set_value(n,"Error/Comment","Levels missing:"+str(miss)+" & extra are:"+str(extra))
  except:
    report_df.set_value(n,"Market",Market)
    report_df.set_value(n,"Quality Check","PROD_LEVEL values check")
    report_df.set_value(n,"Status","Pass")
    report_df.set_value(n,"Error/Comment",'No column named PROD_LEVEL')
    
```

```
def getSchemaSFF(df):
  col = [] 
  for j in range(len(df.columns)):
    if df.columns[j]=="DISPLAY_ORDER":
      dsp_ord=df.select("DISPLAY_ORDER").distinct().rdd.flatMap(lambda x : x).collect()
      col.append((df.columns[j],df.dtypes[j][1],dsp_ord))
    elif df.columns[j]=="HIER_NUM":
      hier_num=df.select("HIER_NUM").distinct().rdd.flatMap(lambda x : x).collect()
      col.append((df.columns[j],df.dtypes[j][1],hier_num))
    elif df.columns[j]=="HIER_LEVEL_NUM":
      hier_l_num=df.select("HIER_LEVEL_NUM").distinct().rdd.flatMap(lambda x : x).collect()
      col.append((df.columns[j],df.dtypes[j][1],hier_l_num))
    else:
      col.append((df.columns[j],df.dtypes[j][1],['NA']))
        

  colrdd = sc.parallelize(col)
  coldf = spark.createDataFrame(colrdd,['Column','Data type','Contents'])
  return coldf
```
